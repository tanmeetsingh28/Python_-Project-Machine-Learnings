```python
#The purpose of scaling is to standardize numerical features, ensuring they have comparable ranges and variances, which is crucial 
#for many machine learning algorithms to perform effectively, preventing features with larger numeric ranges from dominating the
#algorithm's learning process. For example, scaling ensures that both age (ranging from 20 to 60 years) and income 
#(ranging from 30,000 to 100,000 USD) contribute equally to a clustering algorithm like K-Means.
```

```python
# If you have a numerical feature x = [10, 20, 30, 40]:
# Mean μ = (10 + 20 + 30 + 40) / 4 = 25
# Standard deviation σ = sqrt(((10-25)^2 + (20-25)^2 + (30-25)^2 + (40-25)^2) / 4) ≈ 11.18
# Applying StandardScaler transforms x as:
# z = [(10 - 25) / 11.18, (20 - 25) / 11.18, (30 - 25) / 11.18, (40 - 25) / 11.18] ≈ [-1.34, -0.45, 0.45, 1.34]
```

```python
df_scaled.head()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans()
```

```python
tsd = []
for k in range(1,12):
    kmeans = KMeans(n_clusters = k)
    kmeans.fit(df_scaled)
    tsd.append(kmeans.inertia_)
tsd
```

```python
# Create a line plot
sns.lineplot(x = np.arange(1,12), y = tsd)

# Display the plot
plt.show()
```
