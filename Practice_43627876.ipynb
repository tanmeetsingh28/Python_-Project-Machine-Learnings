__Write a function to find the employee with the highest performance score in each department.__

```python
def high_p_e(data):
    top_indices = data.groupby('Department')['Salary'].idxmax()
    top_performers = data.loc[top_indices]
    return top_performers
```

```python
Top_Performers_Data = high_p_e(data)
```

```python
def calculate_years_of_work(joining_date):
    # Ensure joining_date is a datetime object
    joining_date = pd.to_datetime(joining_date)
    # Get the current date
    current_date = pd.Timestamp.now()
    # Calculate the difference in years
    years_of_work = (current_date - joining_date).days / 365.25
    return years_of_work
```

```python
data['years_of_workkkk'] = round(data['JoiningDate'].apply(calculate_years_of_work),0)
```

__Write a function to rank employees within each department based on their 'Salary'.__

```python
def rank_employees_by_salary(data):
    data['Salary_Rank'] = data.groupby('Department')['Salary'].rank(ascending=False, method='dense')
    return data
```

```python
ranked_df = rank_employees_by_salary(data)
```
